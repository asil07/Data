Keywords that searched past 12 month in **South Korea** according to **Google**

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import matplotlib.dates as mdates
```

```python
df_keywords = pd.read_csv("keywords.csv")
df_keywords.Week = pd.to_datetime(df_keywords.Week)
df_keywords.rename(columns={'Week': 'DATE'}, inplace=True)
df_keywords = df_keywords.resample("M", on="DATE").sum()
print(f'Shape of Data: {df_keywords.shape}')
df_keywords
```

```python
plt.figure(figsize=(20, 12))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Number of Search", fontsize=14)
plt.title("Number of Search specific keywords in South Korea past 12 month")

ax1 = plt.gca()

# Locators for ticks
week = mdates.WeekdayLocator()
month = mdates.MonthLocator()
years_fmt = mdates.DateFormatter("%Y-%m")

for column in df_keywords.columns:
  ax1.plot(df_keywords.index, df_keywords[column], label=df_keywords[column].name)
plt.legend(fontsize=14)

ax1.xaxis.set_major_locator(month)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(week)
```
